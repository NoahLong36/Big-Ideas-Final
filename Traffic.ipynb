```python
import pandas as pd
%matplotlib inline



bikedata = "biketraffic.csv"
bikes_data = pd.read_csv(bikedata, index_col = "neighborhood")
bikes_data.head(10)
```

I need to filter out the unnecessary data and focus in on neighborhood and average car traffic.

```python
df = bikes_data[["average_daily_car_traffic"]]
bikes_data.head()
df = df.sort_values(by="neighborhood")
df.head(100)
df = df['average_daily_car_traffic'].sort_values(ascending=True)
df
```

Sorted by least to greatest to find the least amount of traffic in neighborhoods, let's plot it now.

```python
df.head(20).plot(kind = "bar")
```

As a result, Central Northside would be the best place to ride a bike since it has the least amount of daily car traffic.
