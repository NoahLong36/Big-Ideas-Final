```python
import pandas as pd
%matplotlib inline



obstructiondata = "Obstructions.csv"
obs_data = pd.read_csv(obstructiondata, index_col = "id")
obs_data.head(10)
```

I need to filter out the unnecessary data. I just need the neighborhood, and what the permit is for.

```python
df = obs_data[["neighborhood","type"]]
df.head(10)
```

This is helpful, but I should see the count instead of each individual permit.

```python
df = df['neighborhood'].value_counts().sort_values(ascending=True)
df
```

Sorting from least to greatest to find neighborhood with least obstructions.

```python
df.head(20).plot(kind = "bar")
```
